## Additional Statistical Analysis


```python

from scipy.stats import ttest_ind, chi2_contingency

# Example: T-test for sales in different regions
region1_sales = superstore_df[superstore_df['Region'] == 'Central']['Sales']
region2_sales = superstore_df[superstore_df['Region'] == 'East']['Sales']

t_stat, p_value=ttest_ind(region1_sales,region2_sales,equal_var =False)
print(f'T-Stats is:{t_stat}')
print(f'p_value is:{p_value}')
```

```python
#performing chi square test
contingency_table = pd.crosstab(superstore_df['Product Category'],superstore_df['Region'])
chi2_stat, p_value,dof, expected=chi2_contingency(contingency_table)
print(f'Chi-Square test result:{chi2_stat}')
print(f'p_value is:{p_value}')
```

### Sales and profit trend over period of time

```python
# Plotting sales trends over time
superstore_df['Order Date'] = pd.to_datetime(superstore_df['Order Date'])
superstore_df.set_index('Order Date', inplace=True)

# Resample sales data by month
monthly_sales = superstore_df['Sales'].resample('M').sum()

monthly_sales.plot(kind='line', figsize=(12, 6))
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Plotting profit trends over time
profit_over_time = superstore_df['Profit'].resample('M').sum()
profit_over_time.plot(kind='line', figsize=(12, 6))
plt.title('Profit Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Profit')
plt.show()
```

```python
# Identifying top-performing product categories based on total sales and profit
top_sales_cat = superstore_df.groupby('Product Category')['Sales'].sum()
top_profit_cat = superstore_df.groupby('Product Category')['Profit'].sum()

print(top_sales_cat)

# Plotting total sales by product category
sns.barplot(x=top_sales_cat.index, y=top_sales_cat.values)
plt.title('Top performing product by sales')
plt.xlabel('Product Category')
plt.ylabel('Profit')
plt.show()
```

```python
# Plotting total profit by product category
print(top_profit_cat)

sns.barplot(x=top_profit_cat.index, y=top_profit_cat.values)
plt.title('Top Profit by Product')
plt.xlabel('Product Category')
plt.ylabel('Profit')
plt.show()
```

```python

# Generating summary statistics for each region
Region_summary = superstore_df.groupby('Region').agg(
    {'Sales': 'sum'},
    {'Profit': 'sum'},
    {'Quantity': 'sum'}
)
# Display summary statistics
print('Region_Summary\n')
print(Region_summary)
```

```python
# Visualizing the correlation between sales and profit
sns.scatterplot(data=superstore_df, x='Sales',y='Profit')
plt.title('correlation between sales and profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()

# Displaying the correlation coefficient
correlation = superstore_df[['Sales','Profit']].corr()
print('Correlation coefficient between sales and profit\n',correlation)
```
